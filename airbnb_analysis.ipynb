Creating a function to display for all **neighbourhoods**

```python
def correl(name):
    dfnbh=nycab.loc[(nycab['neighbourhood_group_cleansed']==name)]
    #dfnbh.tail()
    df2=dfnbh.drop(['neighbourhood_group_cleansed','NeighbourhoodArea','Listings'], axis=1)
    correlation_nbh = df2.corr()
    plt.figure(figsize=(10,10))
    sns.heatmap(correlation_nbh,cbar = True, square = True, cmap= 'Blues', annot = True, fmt = '.1f')
```

```python
correl('Bronx')
```
